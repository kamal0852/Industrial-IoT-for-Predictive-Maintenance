```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

df = pd.read_excel("survey data.xlsx")

print(df.head())-
```

```python
print(df.shape)
print(df.columns)
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns

df[numeric_cols] = df[numeric_cols].fillna(
    df[numeric_cols].mean()
)

print(df.isnull().sum())
```

```python
summary = df[numeric_cols].describe()
summary
```

```python
question_means = df[numeric_cols].mean()
print(question_means)
```

```python
correlation = df[numeric_cols].corr()
correlation
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# correlation matrix
correlation = df[numeric_cols].corr()

# bigger figure size
plt.figure(figsize=(16,10))

# heatmap
sns.heatmap(
    correlation,
    annot=False,      # values hide kar do, graph clean dikhega
    cmap='coolwarm',
    linewidths=0.5
)

plt.title("Correlation Heatmap", fontsize=18)

# labels formatting
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(rotation=0, fontsize=8)

plt.tight_layout()
plt.show()
```

```python
question_means.plot(kind='bar')

plt.title("Average Survey Responses")
plt.xlabel("Questions")
plt.ylabel("Mean Score")
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

obj1_cols = df.columns[6:11]

titles = [
    "Real-Time Machine Monitoring",
    "Reduction of Machine Downtime",
    "Maintenance Accuracy Through IoT Sensors",
    "Early Identification of Machine Faults",
    "Effectiveness of Predictive Maintenance"
]

fig, axes = plt.subplots(2, 3, figsize=(18,10))
axes = axes.flatten()

for i, col in enumerate(obj1_cols):

    counts = df[col].value_counts().sort_index()

    counts.plot(
        kind='bar',
        ax=axes[i]
    )

    axes[i].set_title(
        titles[i],
        fontsize=12,
        fontweight='bold'
    )

    axes[i].set_xlabel(
        "Likert Scale Response",
        fontsize=10
    )

    axes[i].set_ylabel(
        "Number of Respondents",
        fontsize=10
    )

    axes[i].tick_params(axis='x', rotation=0)

fig.delaxes(axes[5])

plt.suptitle(
    "Objective 1: Perceived Benefits of Industrial IoT and Predictive Maintenance",
    fontsize=16,
    fontweight='bold'
)

plt.tight_layout()
plt.show()
```

```python
obj2_cols = df.columns[11:15]

titles = [
    "Reduction in Operational Costs",
    "Use of Industrial IoT Technologies",
    "Use of Predictive Maintenance Strategies",
    "Employee Training and Readiness"
]

fig, axes = plt.subplots(2,2, figsize=(16,10))
axes = axes.flatten()

for i, col in enumerate(obj2_cols):

    counts = df[col].value_counts().sort_index()

    counts.plot(
        kind='pie',
        ax=axes[i]
    )

    axes[i].set_title(
        titles[i],
        fontsize=12,
        fontweight='bold'
    )

    axes[i].set_xlabel(
        "Number of Respondents",
        fontsize=10
    )

    axes[i].set_ylabel(
        "Likert Scale Response",
        fontsize=10
    )

plt.suptitle(
    "Objective 2: Organizational Adoption of Industrial IoT and Predictive Maintenance",
    fontsize=16,
    fontweight='bold'
)

plt.tight_layout()
plt.show()
```

```python
obj3_cols = df.columns[15:19]

titles = [
    "Operational Efficiency Improvement",
    "Equipment Reliability Improvement",
    "Data-Driven Maintenance Decisions",
    "AI-Based Maintenance Performance"
]

fig, axes = plt.subplots(2,2, figsize=(16,10))
axes = axes.flatten()

for i, col in enumerate(obj3_cols):

    counts = df[col].value_counts().sort_index()

    counts.plot(
        kind='line',
        marker='o',
        linewidth=2,
        ax=axes[i]
    )

    axes[i].set_title(
        titles[i],
        fontsize=12,
        fontweight='bold'
    )

    axes[i].set_xlabel(
        "Likert Scale Response",
        fontsize=10
    )

    axes[i].set_ylabel(
        "Number of Respondents",
        fontsize=10
    )

    axes[i].grid(True)

plt.suptitle(
    "Objective 3: Impact of Industrial IoT on Operational Performance",
    fontsize=16,
    fontweight='bold'
)

plt.tight_layout()
plt.show()
```

```python
obj4_cols = df.columns[19:23]

titles = [
    "Benefits vs Implementation Cost",
    "Implementation Cost Barrier",
    "Data Security Concerns",
    "Lack of Skilled Workforce",

]

fig, axes = plt.subplots(2, 2, figsize=(20,10))
axes = axes.flatten()

for i, col in enumerate(obj4_cols):

    counts = df[col].value_counts()

    # Last 2 questions (open-ended)
    if i >= 4:

        counts.head(10).plot(
            kind='pie',
            ax=axes[i]
        )

        axes[i].set_xlabel("Frequency")
        axes[i].set_ylabel("Responses")

    # Likert Questions
    else:

        counts.sort_index().plot(
            kind='bar',
            ax=axes[i]
        )

        axes[i].set_xlabel("Likert Scale Response")
        axes[i].set_ylabel("Number of Respondents")

    axes[i].set_title(
        titles[i],
        fontsize=11,
        fontweight='bold'
    )

plt.suptitle(
    "Objective 4: Challenges and Recommendations for Industrial IoT Adoption",
    fontsize=16,
    fontweight='bold'
)

plt.tight_layout()
plt.show()
```
